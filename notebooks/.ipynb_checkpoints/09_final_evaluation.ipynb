# Final Model Evaluation & Comparison

Comprehensive evaluation of all three recommendation approaches:
- **Content-Based** (TF-IDF + Cosine Similarity)
- **Collaborative Filtering** (SVD Matrix Factorization)
- **Hybrid** (Weighted combination)

**Evaluation Metrics:**
- NDCG@k (ranking quality)
- Precision@k (relevance)
- Coverage (catalog utilization)
- Diversity (recommendation variety)
- Explanation quality samples

```python
# Import libraries
import sys
sys.path.append('..')

import joblib
import pandas as pd
import numpy as np
from pathlib import Path
import json
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics.pairwise import cosine_similarity

# Import our recommender engine
from app.recommender import RecommendationEngine

# Set style
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)

print("✓ Libraries loaded")
```

## 1. Load Models and Data

```python
# Initialize recommender engine (loads all models)
engine = RecommendationEngine()
engine.load_models()

print(f"\n✓ Models loaded:")
print(f"  - TF-IDF vectorizer: {engine.tfidf_vectorizer is not None}")
print(f"  - TF-IDF matrix shape: {engine.tfidf_matrix.shape if engine.tfidf_matrix is not None else 'N/A'}")
print(f"  - CF model loaded: {engine.cf_model is not None}")
print(f"  - Programs catalog: {len(engine.programs_df)} courses")
```

```python
# Load test data
users_df = pd.read_csv('../data/processed/synth_users.csv')
interactions_df = pd.read_csv('../data/processed/user_course_enrollments.csv')

print(f"Test Data:")
print(f"  - Users: {len(users_df)}")
print(f"  - Interactions: {len(interactions_df)}")
print(f"  - Avg interactions per user: {len(interactions_df) / len(users_df):.2f}")
```

## 2. Helper Functions for Evaluation

```python
def ndcg_at_k(relevance_scores, k):
    """
    Calculate NDCG at k.
    relevance_scores: list of relevance scores in recommended order
    """
    relevance_scores = np.array(relevance_scores[:k])
    if len(relevance_scores) == 0:
        return 0.0
    
    # DCG
    dcg = relevance_scores[0] + np.sum(relevance_scores[1:] / np.log2(np.arange(2, len(relevance_scores) + 1)))
    
    # IDCG (ideal DCG with perfect ranking)
    ideal_scores = sorted(relevance_scores, reverse=True)
    idcg = ideal_scores[0] + np.sum(ideal_scores[1:] / np.log2(np.arange(2, len(ideal_scores) + 1)))
    
    return dcg / idcg if idcg > 0 else 0.0

def precision_at_k(relevance_scores, k):
    """
    Calculate Precision at k.
    Precision@k = (# relevant items in top-k) / k
    """
    relevance_scores = np.array(relevance_scores[:k])
    if len(relevance_scores) == 0:
        return 0.0
    return np.sum(relevance_scores >= 4) / k  # 4+ rating is "relevant"

def calculate_coverage(recommendations_list, total_items):
    """
    Calculate catalog coverage (% of unique items recommended).
    """
    unique_items = set()
    for recs in recommendations_list:
        unique_items.update([r[0] for r in recs])  # r[0] is program_id
    return len(unique_items) / total_items

def calculate_diversity(recommendations_list):
    """
    Calculate average pairwise dissimilarity in recommendations.
    """
    diversities = []
    for recs in recommendations_list:
        if len(recs) < 2:
            continue
        # Simple diversity: count unique items / total items
        unique = len(set([r[0] for r in recs]))
        diversities.append(unique / len(recs))
    return np.mean(diversities) if diversities else 0.0

print("✓ Evaluation functions defined")
```

## 3. Sample Test Users

```python
# Select 50 random test users for evaluation
np.random.seed(42)
test_users = users_df.sample(n=min(50, len(users_df)), random_state=42)

print(f"Selected {len(test_users)} test users for evaluation")
print(f"\nSample user profiles:")
print(test_users[['user_id', 'skills']].head(3))
```

## 4. Evaluate Each Model

```python
# Evaluation settings
K_VALUES = [5, 10, 20]
results = {
    'content-based': {'ndcg': {k: [] for k in K_VALUES}, 'precision': {k: [] for k in K_VALUES}, 'recs': []},
    'collaborative': {'ndcg': {k: [] for k in K_VALUES}, 'precision': {k: [] for k in K_VALUES}, 'recs': []},
    'hybrid': {'ndcg': {k: [] for k in K_VALUES}, 'precision': {k: [] for k in K_VALUES}, 'recs': []}
}

print("Evaluating models on test users...")
print("="*60)

for _, user in test_users.iterrows():
    user_id = user['user_id']
    user_interests = user['skills']
    
    # Get user's actual interactions (ground truth)
    user_interactions = interactions_df[interactions_df['user_id'] == user_id]
    actual_courses = set(user_interactions['course_index'].astype(str).tolist())
    
    if not actual_courses:
        continue  # Skip users with no interactions
    
    # 1. Content-Based Recommendations
    try:
        content_recs = engine.content_based_recommendations(user_interests, k=20)
        results['content-based']['recs'].append(content_recs)
        
        relevance = [5 if str(rec[0]) in actual_courses else 0 for rec in content_recs]
        for k in K_VALUES:
            results['content-based']['ndcg'][k].append(ndcg_at_k(relevance, k))
            results['content-based']['precision'][k].append(precision_at_k(relevance, k))
    except Exception as e:
        print(f"Content-based error for user {user_id}: {e}")
    
    # 2. Collaborative Filtering Recommendations
    try:
        collab_recs = engine.collaborative_recommendations(user_id, k=20)
        if not collab_recs:
            collab_recs = engine.new_user_collaborative_recommendations(user_interests, k=20)
        
        results['collaborative']['recs'].append(collab_recs)
        
        relevance = [5 if str(rec[0]) in actual_courses else 0 for rec in collab_recs]
        for k in K_VALUES:
            results['collaborative']['ndcg'][k].append(ndcg_at_k(relevance, k))
            results['collaborative']['precision'][k].append(precision_at_k(relevance, k))
    except Exception as e:
        print(f"Collaborative error for user {user_id}: {e}")
    
    # 3. Hybrid Recommendations
    try:
        hybrid_recs_raw = engine.hybrid_recommendations(user_interests, user_id=user_id, k=20)
        hybrid_recs = [(r['program_id'], r['score'], r['explanation']) for r in hybrid_recs_raw]
        results['hybrid']['recs'].append(hybrid_recs)
        
        relevance = [5 if str(rec[0]) in actual_courses else 0 for rec in hybrid_recs]
        for k in K_VALUES:
            results['hybrid']['ndcg'][k].append(ndcg_at_k(relevance, k))
            results['hybrid']['precision'][k].append(precision_at_k(relevance, k))
    except Exception as e:
        print(f"Hybrid error for user {user_id}: {e}")

print(f"\n✓ Evaluation complete for {len(test_users)} users")
```

## 5. Calculate Summary Statistics

```python
# Calculate averages and coverage/diversity
summary = []
total_courses = len(engine.programs_df)

for model_name in ['content-based', 'collaborative', 'hybrid']:
    coverage = calculate_coverage(results[model_name]['recs'], total_courses)
    diversity = calculate_diversity(results[model_name]['recs'])
    
    for k in K_VALUES:
        avg_ndcg = np.mean(results[model_name]['ndcg'][k]) if results[model_name]['ndcg'][k] else 0
        avg_precision = np.mean(results[model_name]['precision'][k]) if results[model_name]['precision'][k] else 0
        
        summary.append({
            'Model': model_name,
            'k': k,
            'NDCG@k': avg_ndcg,
            'Precision@k': avg_precision,
            'Coverage': coverage,
            'Diversity': diversity
        })

summary_df = pd.DataFrame(summary)
print("\n" + "="*80)
print("FINAL EVALUATION RESULTS")
print("="*80)
print(summary_df.to_string(index=False))
print("="*80)
```

## 6. Visualization: Model Comparison

```python
# Plot NDCG comparison
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# NDCG plot
for model in ['content-based', 'collaborative', 'hybrid']:
    model_data = summary_df[summary_df['Model'] == model]
    axes[0].plot(model_data['k'], model_data['NDCG@k'], marker='o', label=model, linewidth=2)

axes[0].set_xlabel('k (Top-k Recommendations)', fontsize=12)
axes[0].set_ylabel('NDCG@k', fontsize=12)
axes[0].set_title('Ranking Quality: NDCG@k', fontsize=14, fontweight='bold')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# Precision plot
for model in ['content-based', 'collaborative', 'hybrid']:
    model_data = summary_df[summary_df['Model'] == model]
    axes[1].plot(model_data['k'], model_data['Precision@k'], marker='s', label=model, linewidth=2)

axes[1].set_xlabel('k (Top-k Recommendations)', fontsize=12)
axes[1].set_ylabel('Precision@k', fontsize=12)
axes[1].set_title('Relevance: Precision@k', fontsize=14, fontweight='bold')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('../models/evaluation_comparison.png', dpi=150, bbox_inches='tight')
plt.show()

print("✓ Plots saved to models/evaluation_comparison.png")
```

```python
# Coverage and Diversity comparison
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Get unique coverage/diversity per model (same for all k)
model_summary = summary_df.groupby('Model').first()[['Coverage', 'Diversity']].reset_index()

# Coverage
axes[0].bar(model_summary['Model'], model_summary['Coverage'], color=['#3498db', '#e74c3c', '#2ecc71'])
axes[0].set_ylabel('Coverage (% of Catalog)', fontsize=12)
axes[0].set_title('Catalog Coverage', fontsize=14, fontweight='bold')
axes[0].set_ylim([0, 1])

# Diversity
axes[1].bar(model_summary['Model'], model_summary['Diversity'], color=['#3498db', '#e74c3c', '#2ecc71'])
axes[1].set_ylabel('Diversity Score', fontsize=12)
axes[1].set_title('Recommendation Diversity', fontsize=14, fontweight='bold')
axes[1].set_ylim([0, 1])

plt.tight_layout()
plt.show()
```
